```python
df_gpa=df_gpa.drop(['Counseling Attendance','Stress Coping Mechanisms','Family Mental Health History','Medical Condition'],axis=1)
```

```python
df_gpa.head()
```

```python
df_gpa=pd.get_dummies(df_gpa,columns=['Gender'],drop_first=True)
df_gpa=df_gpa.astype(int)
```

```python
df_gpa.head()
```

```python
y=df_gpa['Academic Performance (GPA)']
X=df_gpa.drop('Academic Performance (GPA)',axis=1)
```

<h1>Train - Test Split</h1>

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_test
```

```python
X_train
```

<h1>Standardization</h1>

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import pickle 
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
print(X_test.dtype)
```

<h1>Training the model</h1>

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model_predict=model.predict(X_test)
```

```python
model_predict
```

```python
plt.scatter(y_test,model_predict)
```

```python
residuals=y_test-model_predict
```

```python
residuals
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print(mean_absolute_error(y_test,model_predict))
print(mean_squared_error(y_test,model_predict))
print(np.sqrt(mean_squared_error(y_test,model_predict)))
```
